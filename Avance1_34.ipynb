````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

### **Estructura de los datos**

Se incluye una descripción general de la forma y los tipos de datos. Se obtienen estadísticas descriptivas para todas las variables del conjunto. En variables categóricas se determina la frecuencia de las clases. Se identifican valores faltantes.

````python
# Cargamos el conjunto de datos en un DataFrame de pandas
data = pd.read_excel('descriptivos oracle.xlsx')
````

````python
# Descripción general de la forma y los tipos de datos
data.info()
````

````python
# Generamos las estadísticas descriptivas
data.describe()
````

````python
# Frecuencias de clases para variables categóricas
data['Género'].value_counts()
````

````python
# Identificar valores faltantes
data.isnull().sum()
````

### **Análisis univariante**

Se emplean técnicas de visualización (histogramas, boxplot, gráficos de barras) en función del tipo de datos y la escala de medición.

- **Gráfico de barras para la variable categórica "Género"**: Muestra la frecuencia de cada género en el conjunto de datos. <br>
- **Gráfico de barras para la variable categórica "Desempeño"**: Muestra la frecuencia de cada nivel de desempeño en el conjunto de datos. <br>
- **Gráfico de barras para la variable categórica "Profesión"**: Muestra la frecuencia de cada profesión en el conjunto de datos. Se rota la etiqueta de las profesiones para mejorar la legibilidad. <br>

````python
# Visualizaciones para variables categóricas

# Gráfico de barras para Género
plt.bar(data["Género"].value_counts().index, data["Género"].value_counts().values)
plt.xlabel("Género")
plt.ylabel("Frecuencia")
plt.title("Frecuencia por género")
plt.show()
````

````python
# Gráfico de barras para Desempeño
plt.bar(data["Desempeño"].value_counts().index, data["Desempeño"].value_counts().values)
plt.xlabel("Desempeño")
plt.ylabel("Frecuencia")
plt.title("Frecuencia por desempeño")
plt.show()
````

````python
# Gráfico de barras para Profesión
plt.bar(data["Profesión"].value_counts().index, data["Profesión"].value_counts().values)
plt.xlabel("Profesión")
plt.ylabel("Frecuencia")
plt.title("Frecuencia por profesión")
plt.xticks(rotation=45, ha='right')  #Se rota la etiqueta de las profesiones para mejorar la legibilidad.
plt.show()
````

### **Análisis bi/multivariante**

Se efectúan análisis de correlación (cuantitativos y gráficos) para explorar las relaciones entre las variables importantes del conjunto de datos.



El análisis de correlación no es adecuado para datos categóricos y descriptivos.

En su lugar, podemos realizar lo siguiente:

````python
#Análisis de frecuencia para entender la distribución de cada variable:
profession_counts = data['Profesión'].value_counts()
print("Frecuencia por profesión:")
print(profession_counts)
````

````python
gender_counts = data['Género'].value_counts()
print("\nFrecuencia por género:")
print(gender_counts)
````

````python
performance_counts = data['Desempeño'].value_counts()
print("\nFrecuencia por desempeño:")
print(performance_counts)
````

Análisis descriptivo para resumir las características de cada variable:

````python
profession_descriptives = data.groupby('Profesión')['Descriptivo'].nunique()
print("\nDescriptivos por profesión:")
print(profession_descriptives)
````

````python
gender_descriptives = data.groupby('Género')['Descriptivo'].nunique()
print("\nDescriptivos por género:")
print(gender_descriptives)
````

````python
performance_descriptives = data.groupby('Desempeño')['Descriptivo'].nunique()
print("\nDescriptivos por desempeño:")
print(performance_descriptives)
````

Visualizar las distribuciones de frecuencia utilizando gráficos de barras.

````python
plt.bar(profession_counts.index, profession_counts.values)
plt.xlabel("Profesión")
plt.ylabel("Frecuencia")
plt.title("Frecuencia por profesión")
plt.xticks(rotation=45, ha='right')  # Rotate profession labels for better readability
plt.show()
````

````python
plt.bar(gender_counts.index, gender_counts.values)
plt.xlabel("Género")
plt.ylabel("Frecuencia")
plt.title("Frecuencia por género")
plt.show()
````

````python
plt.bar(performance_counts.index, performance_counts.values)
plt.xlabel("Desempeño")
plt.ylabel("Frecuencia")
plt.title("Frecuencia por desempeño")
plt.show()
````
