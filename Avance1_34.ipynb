````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

### **Estructura de los datos**

Se incluye una descripcin general de la forma y los tipos de datos. Se obtienen estadsticas descriptivas para todas las variables del conjunto. En variables categricas se determina la frecuencia de las clases. Se identifican valores faltantes.

````python
# Cargamos el conjunto de datos en un DataFrame de pandas
data = pd.read_excel('descriptivos oracle.xlsx')
````

````python
# Descripcin general de la forma y los tipos de datos
data.info()
````

````python
# Generamos las estadsticas descriptivas
data.describe()
````

````python
# Frecuencias de clases para variables categricas
data['Gnero'].value_counts()
````

````python
# Identificar valores faltantes
data.isnull().sum()
````

### **Anlisis univariante**

Se emplean tcnicas de visualizacin (histogramas, boxplot, grficos de barras) en funcin del tipo de datos y la escala de medicin.

- **Grfico de barras para la variable categrica "Gnero"**: Muestra la frecuencia de cada gnero en el conjunto de datos. <br>
- **Grfico de barras para la variable categrica "Desempeo"**: Muestra la frecuencia de cada nivel de desempeo en el conjunto de datos. <br>
- **Grfico de barras para la variable categrica "Profesin"**: Muestra la frecuencia de cada profesin en el conjunto de datos. Se rota la etiqueta de las profesiones para mejorar la legibilidad. <br>

````python
# Visualizaciones para variables categricas

# Grfico de barras para Gnero
plt.bar(data["Gnero"].value_counts().index, data["Gnero"].value_counts().values)
plt.xlabel("Gnero")
plt.ylabel("Frecuencia")
plt.title("Frecuencia por gnero")
plt.show()
````

````python
# Grfico de barras para Desempeo
plt.bar(data["Desempeo"].value_counts().index, data["Desempeo"].value_counts().values)
plt.xlabel("Desempeo")
plt.ylabel("Frecuencia")
plt.title("Frecuencia por desempeo")
plt.show()
````

````python
# Grfico de barras para Profesin
plt.bar(data["Profesin"].value_counts().index, data["Profesin"].value_counts().values)
plt.xlabel("Profesin")
plt.ylabel("Frecuencia")
plt.title("Frecuencia por profesin")
plt.xticks(rotation=45, ha='right')  #Se rota la etiqueta de las profesiones para mejorar la legibilidad.
plt.show()
````

### **Anlisis bi/multivariante**

Se efectan anlisis de correlacin (cuantitativos y grficos) para explorar las relaciones entre las variables importantes del conjunto de datos.



El anlisis de correlacin no es adecuado para datos categricos y descriptivos.

En su lugar, podemos realizar lo siguiente:

````python
#Anlisis de frecuencia para entender la distribucin de cada variable:
profession_counts = data['Profesin'].value_counts()
print("Frecuencia por profesin:")
print(profession_counts)
````

````python
gender_counts = data['Gnero'].value_counts()
print("\nFrecuencia por gnero:")
print(gender_counts)
````

````python
performance_counts = data['Desempeo'].value_counts()
print("\nFrecuencia por desempeo:")
print(performance_counts)
````

Anlisis descriptivo para resumir las caractersticas de cada variable:

````python
profession_descriptives = data.groupby('Profesin')['Descriptivo'].nunique()
print("\nDescriptivos por profesin:")
print(profession_descriptives)
````

````python
gender_descriptives = data.groupby('Gnero')['Descriptivo'].nunique()
print("\nDescriptivos por gnero:")
print(gender_descriptives)
````

````python
performance_descriptives = data.groupby('Desempeo')['Descriptivo'].nunique()
print("\nDescriptivos por desempeo:")
print(performance_descriptives)
````

Visualizar las distribuciones de frecuencia utilizando grficos de barras.

````python
plt.bar(profession_counts.index, profession_counts.values)
plt.xlabel("Profesin")
plt.ylabel("Frecuencia")
plt.title("Frecuencia por profesin")
plt.xticks(rotation=45, ha='right')  # Rotate profession labels for better readability
plt.show()
````

````python
plt.bar(gender_counts.index, gender_counts.values)
plt.xlabel("Gnero")
plt.ylabel("Frecuencia")
plt.title("Frecuencia por gnero")
plt.show()
````

````python
plt.bar(performance_counts.index, performance_counts.values)
plt.xlabel("Desempeo")
plt.ylabel("Frecuencia")
plt.title("Frecuencia por desempeo")
plt.show()
````
