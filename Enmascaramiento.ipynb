````python
def enmascarar(descriptivo, genero):
  prompt = 'El siguiente texto "' + descriptivo + '" que corresponde a un candidato del genero "' + genero + '", me puedes ayudar a encontrar las palabras clave ' \
           'que identifican el gnero y generar un nuevo texto donde sustituyas dichas palabras clave que identifican el gnero, por texto enmascarado'

  completion = client.chat.completions.create(
    model="gpt-3.5-turbo",
      messages=[
        {"role": "system", "content": "Supongamos que eres experto en enmascarar informacin."},
        {"role": "user", "content":  prompt}
      ]
  )

  return completion.choices[0].message.content
````
